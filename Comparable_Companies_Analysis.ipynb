# Comparable Companies Analysis - United Healthcare Group

```python
import requests 
import pandas as pd
import time

import requests

# Set the criteria for API
sample_api = ' '
sector = 'Healthcare'
industry = 'Healthcare Plans'

# Grab the comparable companies based on above criteia 
companies = requests.get(f'https://fmpcloud.io/api/v3/stock-screener?sector={sector}&industry={industry}&marketCapMoreThan=1000000000&limit=8&apikey={demo}')
# Convert into json style 
companies = companies.json()

technological_companies = []

for item in companies:
  technological_companies.append(item['symbol'])
print(technological_companies)

# Selectively retrieve only comnpanies that are in the US 
interestedIndices = [0, 1, 2, 3, 4, 5, 7]

metrics = {}
tryCount = 0
# Retrieve various metrics based on the above indices
for item in [technological_companies[index] for index in interestedIndices]:
    tryCount = 0
    while True:
        time.sleep(1)
        try:
            tryCount = tryCount + 1
            print("fetching " + item)
            metrics[item] = {}
        
            # 1. Company Name
            company_name = requests.get(f'https://fmpcloud.io/api/v3/profile/{item}?apikey={demo}')
            company_name = company_name.json()
            company_name = company_name[0]['companyName']
            metrics[item]['Name'] = company_name
            
            # 2. Stock Price 
            stock_price = requests.get(f'https://fmpcloud.io/api/v3/profile/{item}?apikey={demo}')
            stock_price = stock_price.json()
            stock_price = stock_price[0]['price']
            metrics[item]['Price'] = stock_price
            
            # 3. Market Capitalization 
            MarketCapitazation = requests.get(f'https://fmpcloud.io/api/v3/market-capitalization/{item}?apikey={demo}')
            MarketCapitazation = MarketCapitazation.json()
            MarketCapitazation = MarketCapitazation[0]['marketCap']
            metrics[item]['Market_Cap'] = MarketCapitazation
            
            # 4. Total Enterprise Value
            TEV = requests.get(f'https://financialmodelingprep.com/api/v3/enterprise-values/{item}?apikey={demo}')
            TEV = TEV.json()
            TEV = TEV[0]['enterpriseValue']
            metrics[item]['TEV'] = TEV
            
            # 5. Sales 
            Sales = requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/{item}?apikey={demo}')
            Sales = Sales.json()
            Sales = Sales[0]['revenue']
            metrics[item]['Sales'] = Sales
              
            # 6. EBITDA
            EBITDA = requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/{item}?apikey={demo}')
            EBITDA = EBITDA.json()
            EBITDA = EBITDA[0]['ebitda']
            metrics[item]['EBITDA'] = EBITDA
            
            # 7. EBIT
            EBIT = requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/{item}?apikey={demo}')
            EBIT = EBIT.json()
            EBIT = EBIT[0]['operatingIncome']
            metrics[item]['EBIT'] = EBIT
            
            
            # 8. PE Ratio   
            PE = requests.get(f'https://financialmodelingprep.com/api/v3/key-metrics/{item}?&apikey={demo}')
            PE = PE.json()
            PE = PE[0]['peRatio']
            metrics[item]['PE'] = PE
            
            # 9. EV/Sales
            metrics[item]['EV/Sales'] = TEV / Sales
            
            # 10. EV/EBITDA
            metrics[item]['EV/EBITDA'] = TEV / EBITDA
            
            # 11. EV/EBIT
            metrics[item]['EV/EBIT'] = TEV / EBIT           

        except:
          if tryCount > 3:
            break
          print("failed " + item)
          continue
        break

print(metrics)

metrics_df = pd.DataFrame.from_dict(metrics, orient='index')
# metrics_df = metrics_df.T
metrics_df
```
